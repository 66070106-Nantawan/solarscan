# 🔍 SolarScan — EDA
วิเคราะห์ dataset ก่อน train

```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

DATA_DIR = Path('../data/raw')
CLASSES  = ['solar', 'no_solar']
```

```python
# นับจำนวนภาพแต่ละ class
for cls in CLASSES:
    imgs = list((DATA_DIR / cls).glob('*.jpg')) + list((DATA_DIR / cls).glob('*.png'))
    print(f'{cls}: {len(imgs)} images')
```

```python
# แสดงตัวอย่างภาพ
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for row, cls in enumerate(CLASSES):
    imgs = list((DATA_DIR / cls).glob('*.jpg'))[:4]
    for col, img_path in enumerate(imgs):
        axes[row][col].imshow(Image.open(img_path))
        axes[row][col].set_title(cls)
        axes[row][col].axis('off')
plt.suptitle('Dataset Samples', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# เช็ค class balance
counts = []
for cls in CLASSES:
    imgs = list((DATA_DIR / cls).glob('*.jpg')) + list((DATA_DIR / cls).glob('*.png'))
    counts.append(len(imgs))

plt.figure(figsize=(6, 4))
plt.bar(CLASSES, counts, color=['#2196F3', '#FF9800'])
plt.title('Class Distribution')
plt.ylabel('จำนวนภาพ')
for i, v in enumerate(counts):
    plt.text(i, v + 1, str(v), ha='center')
plt.show()

ratio = counts[0] / counts[1]
print(f'Class ratio (solar:no_solar) = {ratio:.2f}')
if ratio > 1.5 or ratio < 0.67:
    print('⚠️  Dataset ไม่ balance — ควรใช้ class_weight หรือ oversample')
else:
    print('✅ Dataset balance ดี')
```

```python
# เช็คขนาดภาพ
sizes = []
for cls in CLASSES:
    for img_path in (DATA_DIR / cls).glob('*.jpg'):
        with Image.open(img_path) as img:
            sizes.append(img.size)

widths  = [s[0] for s in sizes]
heights = [s[1] for s in sizes]
print(f'Width:  min={min(widths)}, max={max(widths)}, mean={np.mean(widths):.0f}')
print(f'Height: min={min(heights)}, max={max(heights)}, mean={np.mean(heights):.0f}')
```
